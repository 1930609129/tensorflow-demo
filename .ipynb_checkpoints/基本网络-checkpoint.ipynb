```python
x_train=x[:40]
y_train=y[:40]
x_test=x[40:]
y_test=y[40:]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x_train,y_train,c='b',label='train data')
plt.scatter(x_test,y_test,c='g',label='test data')
plt.legend()
```

```python
tf.random.set_seed(42)
model=Sequential([
#     layers.Dense(10,input_shape=[1],name='input_layer'),
    layers.Dense(1,input_shape=[1],name='output_layer')
],name='model_1')

model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(learning_rate=0.0001),metrics=['mae'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=100,verbose=0)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
def plot_predictions(train_data=x_train,train_labels=y_train,
                     test_data=x_test,test_labels=y_test,
                    predictions=y_pred):
    plt.figure(figsize=(8,5))
    
    plt.scatter(train_data,train_labels,c='b',label='Train data'),
    plt.scatter(test_data,test_labels,c='g',label='Test data')
    
    plt.scatter(test_data,predictions,c='r',label='predictions')
    plt.legend()
```

```python
plot_predictions()
```

```python
model.evaluate(x_test,y_test)
```

```python
def mae(y_true,y_pred):
    return tf.metrics.mean_absolute_error(y_true=y_true,y_pred=tf.squeeze(y_pred))
def mse(y_true,y_pred):
    return tf.metrics.mean_squared_error(y_true=y_true,y_pred=tf.squeeze(y_pred))
```

```python
import pandas as pd
insurance=pd.read_csv('insurance.csv')
insurance
```

```python
insurance['smoker']
```

```python
insurance_one_hot=pd.get_dummies(insurance)
insurance_one_hot.head()
```

```python
x=insurance_one_hot.drop('charges',axis=1)
y=insurance_one_hot['charges']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
tf.random.set_seed(42)
insurance_model=Sequential([
    layers.Dense(10),
    layers.Dense(1)
])
insurance_model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=['mae'])
insurance_model.fit(x_train,y_train,epochs=100)
```

```python
insurance_model.evaluate(x_test,y_test)
```

```python
tf.random.set_seed(42)

insurance_model_2=Sequential([
    layers.Dense(100),
    layers.Dense(10),
    layers.Dense(1)
])
insurance_model_2.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(0.001),metrics=['mae'])
insurance_model_2.fit(x_train,y_train,epochs=100)
```

```python
insurance_model_2.evaluate(x_test,y_test)
```

```python
tf.random.set_seed(42)

insurance_model_3=Sequential([
    layers.Dense(100),
    layers.Dense(10),
    layers.Dense(1)
])

insurance_model_3.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=['mae'])
history=insurance_model_3.fit(x_train,y_train,epochs=200)
```

```python
insurance_model_3.evaluate(x_test,y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel('loss')
plt.xlabel('epochs')
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder

ct=make_column_transformer((MinMaxScaler(),['age','bmi','children']),(OneHotEncoder(handle_unknown='ignore'),
                                                                     ['sex','smoker','region']))
x=insurance.drop('charges',axis=1)
y=insurance['charges']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
ct.fit(x_train)
x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
```

```python
tf.random.set_seed(42)

insurance_model_4=Sequential([
    layers.Dense(100),
    layers.Dense(10),
    layers.Dense(1)
])

insurance_model_4.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=['mae'])
insurance_model_4.fit(x_train_normal,y_train,epochs=100)
```

```python
insurance_model_4.evaluate(x_test_normal,y_test)
```
